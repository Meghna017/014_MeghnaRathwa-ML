```python
import numpy as np
import pandas as pd
```

```python
#Two numpy array of size 3 x 2 and 2 x 3
array1= np.array([[2,4,7],[5,8,4]])
array2= np.array([[8,-6],[9,3],[2,-9]])
print(array1)
print(array2)
```

```python
#randomly initalize that array
array = np.random.rand(3,6)
print(array)
```

```python
#perform matrix multiplacation
mul = np.dot(array1, array2)
print(mul)
```

```python
#perfrom element wise matrix multiplacation
res = [[0 for x in range(len(array1))] for y in range(len(array2[0]))]

for i in range(len(array1)): 
  for j in range(len(array2[0])):
    for k in range(len(array2)):
      res[i][j] += array1[i][k] * array2[k][j]

print(res)
```

```python
#mean of first matrix
mean = np.mean(array1)
print(mean)
```
